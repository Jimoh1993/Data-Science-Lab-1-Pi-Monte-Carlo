```python
import random as r
import math as m
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


inside = 0
total = 200

arr = np.zeros(total)

for i in range(0, total):
  # Generate random x, y in [0, 1].
  x2 = r.random()**2
  y2 = r.random()**2
  if m.sqrt(x2 + y2) <= 1.0:
    inside += 1
  arr[i] = (float(inside) / (i+1)) * 4

print(arr)
```

```python
fig, ax=plt.subplots()
ax.plot(arr, markersize=2)
ax.axhline(y=3.14159265359, color="red")
# ax.axhline(y=np.mean(arr), color="blue")
```

```python
# Covert arr to Data Frame
df = pd.DataFrame(arr).T
df
```
